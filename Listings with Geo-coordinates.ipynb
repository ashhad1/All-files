```python
#count nulls for latitudes and longitudes
df['latitude'].isnull().sum()
df['longitude'].isnull().sum()
```

```python
#Drop rows with zeros in latitude and longitude
df.drop(df[(df.latitude == 0) | (df.longitude==0)].index, inplace=True)
df.head()
```

```python
# count blanks for the building_name fields
df['building_name'].isnull().sum()
```

```python
#drop blanks
df.dropna(subset=['latitude'],inplace='True')
df.dropna(subset=['longitude'],inplace='True')
```

```python
# rounded dfs
new_df=df.round({'latitude': 7, 'longitude': 7})
new_cent=cent.round({'x': 7, 'y': 7})
```

```python
new_df.longitude[13]
```

```python
new_df.latitude[13]
```

```python
# # Check if latitude is a centroid
new_df['latitude'].isin(new_cent.y).head()
```

```python
# Check if longitude is a centroid
new_df['longitude'].isin(new_cent.x).head()
```

```python
## Check matching indicies
indicies=(new_df['latitude'].isin(new_cent.y) & new_df['longitude'].isin(new_cent.x))
np.sum(indicies==True)
```

```python
#all listings
new_df.groupby(['category_l1','category_l2'])['listing_id'].nunique()
```

```python
# listings without a geo-coordinate
df_wo_geo=new_df[indicies]
df_with_geo=new_df[~indicies]
df_wo_geo.groupby(['category_l1','category_l2'])['listing_id'].nunique()
```

```python
#listings with a geo-coordinate
df_with_geo.groupby(['category_l1','category_l2'])['listing_id'].nunique()
```

```python
#listings without a geo-coordinate but have building specified
df_wo_geo[df_wo_geo['building_name'].notnull()].groupby(['category_l1','category_l2'])['listing_id'].nunique()
```
